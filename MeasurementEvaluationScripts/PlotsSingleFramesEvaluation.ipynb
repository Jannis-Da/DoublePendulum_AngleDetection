```python
import pandas as pd
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
pkl_file_path = '../MeasurementData/EinzelbilderEva/EvalOutput.pkl'

with open(pkl_file_path, 'rb') as file:
    pkl_data = pkl.load(file)
    
df = pd.DataFrame(pkl_data)
```

```python
df['abs_vel1'] = abs(df['AngularVel1'])
df_sort_vel1 = df.sort_values('abs_vel1')
df_sort_vel1.plot(x='abs_vel1', y=['abs_err_Angle1'], linestyle='-')
plt.xlabel('absolute angular velocity arm 1 [rad/s]')
plt.ylabel('absolute error angle arm 1 [rad]')
plt.title('Absolute Error vs Velocity Arm 1')
plt.savefig('../MeasurementData/EinzelbilderEva/Plots/AbsoluteErrorVelocityArm1.pdf', format='pdf')
plt.show()

corr_coef_pd = df['abs_vel1'].corr(df['abs_err_Angle1'])
print("Korrelationskoeffizient:", corr_coef_pd)
```

```python
df['abs_vel2'] = abs(df['AngularVel1']) + abs(df['AngularVel2'])
df_sorted_vel2 = df.sort_values('abs_vel2')
df_sorted_vel2.plot(x='abs_vel2', y=['abs_err_Angle2'], linestyle='-')
plt.xlabel('absolute angular velocity arm 2 [rad/s]')
plt.ylabel('absolute error angle arm 2 [rad]')
plt.title('Absolute Error vs Velocity Arm 2')
plt.savefig('../MeasurementData/EinzelbilderEva/Plots/AbsoluteErrorVelocityArm2.pdf', format='pdf')
plt.show()

corr_coef_pd = df['abs_vel2'].corr(df['abs_err_Angle2'])
print("Korrelationskoeffizient:", corr_coef_pd)
```

```python
df['abs_angle2'] = abs(df['gt_Angle2'])
df_sorted_abs_angle2 = df.sort_values('abs_angle2')
df_sorted_abs_angle2.plot(x='abs_angle2', y=['abs_err_Angle2'])
plt.xlabel('absolute angle arm 2 [rad]')
plt.ylabel('absolute error angle arm 2 [rad]')
plt.title('Absolute Error vs Proximity to Frame Edge Arm 2')
plt.savefig('../MeasurementData/EinzelbilderEva/Plots/AbsoluteErrorProximityFrameEdge.pdf', format='pdf')
plt.show()
```
